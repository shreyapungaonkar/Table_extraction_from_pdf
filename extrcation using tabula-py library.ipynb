```python
import pandas as pd
import os
from tabula import read_pdf # import the read_pdf method
```

```python
pdf_file = input("Enter the path of the PDF file: ").strip()  #enter your pdf file path
```

```python
output_folder = "extracted_tables_from_pdf" 
os.makedirs(output_folder, exist_ok=True) #make a new folder to store the extacted tables
```

```python
tabular_data = read_pdf(pdf_file, pages="all")  # read all the pages of pdf
```

```python
tabular_data  #display the tables
```

```python
for i, df in enumerate(tabular_data):   #loop through list of tabluar data extracted and convert them into excel files
    print(df)  # Print each table for verification
    df.to_excel(os.path.join(output_folder, f"table_{i+1}.xlsx"), index=False) #save the excle files into the output folder

print(f"Extracted tables saved in: {output_folder}")
```
